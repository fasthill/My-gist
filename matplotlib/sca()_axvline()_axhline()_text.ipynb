```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin, cos, radians
  
  
fig, ax = plt.subplots(2, 1)
x_degree = np.linspace(0, 360, 360)
x = radians(x_degree)

y1 = sin(x)
y2 = cos(x)

x_tick = [0, 90, 180, 270, 360]
  
plt.sca(ax[0])  # 첫번째 그림판(ax[0])에 작성하기
plt.axhline(y=1, color='r', linestyle='--', linewidth=1) # y=1에 수평선을 긋기
plt.axhline(y=0, color='r', linestyle='--', linewidth=1) # y=0에 수평선을 긋기
plt.axhline(y=-1, color='r', linestyle='--', linewidth=1) # y=-1에 수평선을 긋기
plt.axvline(x=0, color='b', linestyle=':', linewidth=2) # x=0에 수직선을 긋기
plt.axvline(x=180, color='b', linestyle=':', linewidth=2) # x=180에 수직선을 긋기
plt.axvline(x=360, color='b', linestyle=':', linewidth=2) # x=360에 수직선을 긋기
plt.plot(x_degree, y1, label="ax0")
plt.xticks(x_tick)
plt.legend()
  
plt.sca(ax[1]) # 두번째 그림판(ax[1])에 작성하기
plt.text(1, 0.5, r'$\alpha^2 > \beta_5$', fontdict={'size': 12})
plt.plot(x_degree, y2)
  
fig.suptitle("""matplotlib.pyplot.sca()
function Example\n\n""", fontweight ="bold") 
    
plt.show() 
```
