```python
sns.countplot(x=wine_df['quality'])
plt.title('Quality Count', fontsize=16, color='blue')
```

```python
plt.figure(figsize=(16, 16))
corr= wine_df.corr()
sns.heatmap(corr, annot=True, fmt='.2f', linewidths=0.5, cmap='Blues', square=True);
```

```python
def plt_hist(df):
    num_h = 4
    length = len(df.columns)
    num_v = int(length / 4 + 0.5)
    plt.figure(figsize=(3*num_h,3*num_v))
    for i in range(1, length+1):
        plt.subplot(num_v,num_h,i)
        sns.histplot(df.iloc[:,i-1])
    plt.tight_layout()
    plt.show()
    
plt_hist(wine_df)
```
