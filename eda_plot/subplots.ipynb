```python
import numpy as np
import matplotlib.pyplot as plt

# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# 한개의 figure와 한개의 subplot 만들기
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
```

```python
# 두 개의 subplots 만들고 unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)  # y 축을 공동으로 사용
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
# Share a X axis with each column of subplots
plt.subplots(2, 2, sharex='col')

# Share a Y axis with each row of subplots
plt.subplots(2, 2, sharey='row')
```

```python
# Share both X and Y axes with all subplots
plt.subplots(2, 2, sharex='all', sharey='all')
```

```python
# Note that this is the same as
plt.subplots(2, 2, sharex=True, sharey=True)
```

```python
# plt.subplot(221)와 동일하나 좀 더 일반적인 형식
ax1 = plt.subplot(2, 2, 1)  
```

```python
# add a subplot with no frame
ax2 = plt.subplot(222, frameon=False)
```

```python
# add a red subplot that shares the x-axis with ax1
plt.subplot(224, sharex=ax1, facecolor='red')
```

```python
# figure에서 ax2 지우기
plt.delaxes(ax2)
```

```python
plt.subplot(221) # 첫번째 ax를 선택하여 작업하기 (working plot plate)로 설정하기
plt.plot(x, y, '.-')
plt.title('1st Graph')
plt.ylabel('Oscillation')
```

```python
plt.subplot(222) # 두번째 ax를 선택하여 작업하기 (working plot plate)로 설정하기
plt.plot(x, y)
plt.title('2nd Graph')
plt.ylabel('Oscillation')
```

```python
# 한 개의 cell에서 명령어를 수행해야 함.
# 각각의 cell에서 수행하면 위의 예처럼 위치가 어떻게 변하는지 전체의 특성을 볼 수 없음.

plt.subplot(221) # 첫번째 ax를 선택하여 작업하기 (working plot plate)로 설정하기
plt.plot(x, y, '.-')
plt.title('1st Graph')
plt.ylabel('Oscillation')

plt.subplot(222) # 두번째 ax를 선택하여 작업하기 (working plot plate)로 설정하기
plt.plot(x, y)
plt.title('2nd Graph')
plt.ylabel('Oscillation')

plt.subplot(224) # 두번째 ax를 선택하여 작업하기 (working plot plate)로 설정하기
plt.plot(x, y)
plt.title('4th Graph')
plt.ylabel('Oscillation')
```

subplot은 위처럼 각각 동작할 수 있으나, subplots는 한번에 동작함.

```python
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True) # .plot .scatter 등과 같은 cell에서 동작
ax[0,0].plot(x, y)
ax[1,1].scatter(x, y)
```
